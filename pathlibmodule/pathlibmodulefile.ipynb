```python
from pathlib import Path
import os
```

```python
currentdir = Path.cwd();
currentdir

filepath = Path(os.path.join(currentdir, "main.py"))
```

```python
print(currentdir.name);
print(filepath.name)
```

```python
print(currentdir.stem);
print(filepath.stem);
```

```python
print(currentdir.parent);
print(filepath.parent);
```

```python
print(currentdir.exists());
print(filepath.exists());
```

```python
print(currentdir.is_file());
print(filepath.is_file());
```

```python
print(currentdir.is_dir());
print(filepath.is_dir());
```

```python
test_file = os.path.join(currentdir, "test.txt");
test_file_path = Path(test_file).write_text("Hello this is the testing file created using the pathlib module of the python");
```

```python
content_test_file = Path(test_file).read_text();
content_test_file
```

```python
# listing the files of the directory using the pathlib;

[str(file) for file in Path(".").iterdir()]
```

```python
# listing the file with the given extensions using the .glob methodabs;

[str(file) for file in Path(".").glob("*.py")]
```

```python
# recursive searching in the pathlib module using the rglob;

[str(file) for file in Path(".").rglob("*.py")]
```

```python
from pathlib import Path
import os
import datetime
```

```python
currentdir = Path.cwd();
currentdir

filepath = Path(os.path.join(currentdir, "main.py"))
```

```python
print(currentdir.name);
print(filepath.name)
```

```python
print(currentdir.stem);
print(filepath.stem);
```

```python
print(currentdir.parent);
print(filepath.parent);
```

```python
print(currentdir.exists());
print(filepath.exists());
```

```python
print(currentdir.is_file());
print(filepath.is_file());
```

```python
print(currentdir.is_dir());
print(filepath.is_dir());
```

```python
test_file = os.path.join(currentdir, "test.txt");
test_file_path = Path(test_file).write_text("Hello this is the testing file created using the pathlib module of the python");
```

```python
# listing the files of the directory using the pathlib;

[str(file) for file in Path(".").iterdir()]
```

```python
# listing the file with the given extensions using the .glob methodabs;

[str(file) for file in Path(".").glob("*.py")]
```

```python
# recursive searching in the pathlib module using the rglob;

[str(file) for file in Path(".").rglob("*.py")]
```

```python
# Methods to get basic informations about the target directory;

fileinfo = filepath.stat();
directoryinfo = currentdir.stat();

print(f"File Size : {fileinfo.st_size} bytes");
print(f"Modification Time : { datetime.datetime.fromtimestamp(fileinfo.st_mtime) }");
print(f"Creation Time : { datetime.datetime.fromtimestamp(fileinfo.st_ctime) }")

# directory informations;
print("\n\n-----Directory Information-----\n")
print(f"File Size : {directoryinfo.st_size} bytes");
print(f"Modification Time : { datetime.datetime.fromtimestamp(directoryinfo.st_mtime) }");
print(f"Creation Time : { datetime.datetime.fromtimestamp(directoryinfo.st_ctime) }")
```

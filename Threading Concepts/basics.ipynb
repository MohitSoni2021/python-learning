```python
func1Thread = threading.Thread(target=func1, args=(10, 5));
func2Thread = threading.Thread(target=func2, args=(27, 10));

func1Thread.start(); # Output of this function is delayed by 2 second;
func2Thread.start();
```

```python
func1Thread = threading.Thread(target=func1, args=(10, 5));
func2Thread = threading.Thread(target=func2, args=(27, 10));

# If i wanted to func1 to execute 1st and func2 to wait for it, join();

func1Thread.start();
func1Thread.join(); # This tell that after starting of the ffunc1Thread we need to wait before executing the second funtion;

func2Thread.start();
```

## Thread LifeCycle

Created
   ↓
Runnable
   ↓
Running
   ↓
Blocked / Waiting
   ↓
Running
   ↓
Finished (Dead)

```python
# Daemon thread automatically stop when the main program exists;

import threading
import time

stop_event = threading.Event()

def task():
    while True:
        print("Running...")
        time.sleep(1)

t = threading.Thread(target=task, daemon=True)
t.start()

print("Main ends")

stop_event.set();

t.join();
print("Main ends")
```

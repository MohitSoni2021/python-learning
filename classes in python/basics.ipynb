```python
# creating the gamemcp class that define the structure of the game character;
class MCP:
    count = 1;
    def __init__(
        self,
        name,
        health=100,
        armor=0,
        money=0,
        level=1,
        stamina=100,
        weapons=None,
    ):
        if name is None:
            self.name = f"MCP{MCP.count}";
            MCP.count += 1;
        else:
            self.name = name;
            
        self.health = health;
        self.armor = armor;
        self.money = money;
        self.level = level;
        self.stamina = stamina;
        self.weapons = weapons or [];
        self.alive = True;

    def info(self):
        print(f"Name     : {self.name}")
        print(f"Health   : {self.health}")
        print(f"Armor    : {self.armor}")
        print(f"Money    : ${self.money}")
        print(f"Level    : {self.level}")
        print(f"Stamina  : {self.stamina}")
        print(f"Weapons  : {', '.join(self.weapons) if self.weapons else 'None'}")
        print(f"Alive    : {self.alive}")

    def damage(self, amount):
        if self.armor:
            blocked = min(self.armor, amount)
            self.armor -= blocked
            amount -= blocked

        self.health -= amount

        if self.health <= 0:
            self.health = 0
            self.alive = False

    def heal(self, amount):
        if self.alive:
            self.health = min(100, self.health + amount)

    def earn(self, amount):
        self.money += amount

    def spend(self, amount):
        if amount <= self.money:
            self.money -= amount

    def add_weapon(self, weapon):
        self.weapons.append(weapon)

    def remove_weapon(self, weapon):
        if weapon in self.weapons:
            self.weapons.remove(weapon)

    def run(self):
        if self.stamina >= 10:
            self.stamina -= 10

    def rest(self):
        self.stamina = 100

    def level_up(self):
        self.level += 1
        self.health = 100
        self.stamina = 100
```

```python
player = MCP("Mohit", armor=50, money=5000)

player.add_weapon("Pistol")
player.add_weapon("AK-47")

player.damage(30)
player.earn(1000)
player.run()
player.level_up()

player.info()
```

## Polymorphism Concept

class Animal:
    def sound(self):
        pass


class Dog(Animal):
    def sound(self):
        print("Dog says: Woof!")


class Cat(Animal):
    def sound(self):
        print("Cat says: Meow!")


class Cow(Animal):
    def sound(self):
        print("Cow says: Moo!")

```python
animals = [Dog(), Cat(), Cow()]

for animal in animals:
    animal.sound()
```

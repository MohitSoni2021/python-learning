## Inheritance

```python
class Character:
    def __init__(self, name, health=100):
        self.name = name
        self.health = health

    def info(self):
        print(self.name)
        print(self.health)


class Police(Character):
    def __init__(self, badge, **kwargs):
        super().__init__(**kwargs)
        self.badge = badge
```

```python
P1 = Character("Mohit", health=400);
P1.info();
```

```python
p = Police(
    badge="PD101",
    name="Mohit",
    health=200
)

p.info()
print(p.badge)
```

## Multiple Inheritance

```python
class Character:
    def __init__(self, name):
        self.name = name

    def walk(self):
        print(f"{self.name} is walking.")


class Shooter:
    def shoot(self):
        print("Shooting with a gun.")


class Driver:
    def drive(self):
        print("Driving a car.")


class Police(Character, Shooter, Driver):
    def __init__(self, name, badge):
        super().__init__(name)
        self.badge = badge
```

```python
p = Police("John", "PD101")

p.walk()
p.shoot()
p.drive()
```

## Multilevel Inheritance

```python
class Character:
    def walk(self):
        print("Walking")


class Police(Character):
    def arrest(self):
        print("Arresting criminal")


class Inspector(Police):
    def investigate(self):
        print("Investigating case")
```

```python
i = Inspector()

i.walk()
i.arrest()
i.investigate()
```

## Hybrid

```python
class Character:
    def walk(self):
        print("Walking");


class Shooter:
    def shoot(self):
        print("Shoot");


class Police(Character, Shooter):
    def arrest(self):
        print("Arrest");


class Inspector(Police):
    def investigate(self):
        print("Investigate");
```

```python
obj = Inspector();

obj.walk();
obj.shoot();
obj.arrest();
obj.investigate();
```

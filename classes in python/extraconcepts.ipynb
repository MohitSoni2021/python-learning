## Super keyword

```python
class Parent:
    def __init__(self, name, age, university):
        self.name = name;
        self.age = age;
        self.university = university;

    def get_info(self):
        print({
            "name": self.name,
            "age": self.age,
            "university": self.university
        });

class Child(Parent):
    def __init__(self, gender, **kwargs):
        # the super keyword is used to call the constructor of the parent class to invoke the properties and the methods of that class;
        # if we have multiple parent class then we need to specify that class e.g. if we are inherating the parent1 and parent2 the instead of the super() keyword we need to mention parent1().__init__() mentod to invoke the functionality of the parent1 class and same for the parent2 class;
        super().__init__(**kwargs);
        self.gender = gender;

    def __str__(self):
        return str({
            "name": self.name,
            "age": self.age,
            "university": self.university,
            "gender": self.gender
        })

        
```

```python
ObjChild = Child(
    gender="male",
    name="Mohit Soni",
    age=20,
    university="Rai University"
);
```

```python
ObjChild.get_info()
```

```python
print(ObjChild)
```

## Class Variable and Instance variable

```python
class Student:
    # the variable decleared here are the class variables and associated with the class blueprint only;
    reg_no = 1;
    size = 10;
    def __init__(self, name, university):
        # The variables decleared here are the instance variables which update with the instance creation of the class;
        self.name = name;
        self.university = university;
        self.reg_no = Student.reg_no;
        Student.reg_no += 1;

        # The __ used to define the private variables;
        self.__privateValue = 100;

    def __str__(self):
        return str({
            "name": self.name,
            "University": self.university,
            "Registration": self.reg_no
        })

    # Use to define the len(classobject) for the class as per the requirement;
    def __len__(self):
        return self.size;

    # This decorator is used to create methods that is associated to the class itself not the instance of the class;
    # the "cls" take the ref of the class;
    @classmethod
    def currRegisterations(cls):
        print(cls.reg_no)

    # This decorator is used to define the static method which is it means that function present in the class but not associated with the class or the instance of the class;
    @staticmethod
    def staticAdd(a, b):
        print(a+b);

    # The @property is used to define a property for the class;
    # So we don't need to call the function we just use it as the property;
    @property
    def information(self):
        print({"name":self.name, "University": self.university})

    @property
    def getPrivateValue(self):
        return self.__privateValue;

    # The @property.setter is used;
    @getPrivateValue.setter
    def updatePrivateValue(self, newValue):
        self.__privateValue = newValue;
```

```python
s1 = Student(name="Mohit Soni", university="Rai University");
print(s1)
```

```python
s1.currRegisterations();
```

```python
s1.staticAdd(2, 3);
```

```python
s1.information
```

```python
s1.getPrivateValue
```

```python
s1.updatePrivateValue = 20;
```

```python
s1.getPrivateValue
```

## Abstraction in class

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    # An abstract method is a method that is declared but has no implementation in a base (abstract) class. It acts as a contract, requiring every subclass to provide its own implementation.
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):

    def sound(self):
        print("Bark")
```

```python
animalDog = Dog();
animalDog.sound();
```

```python
# Name Mangling, it is used to access the private variable of the class, as per the concepts we can't but still there is a method to access it outside the function;

class A:
    def __init__(self):
        self.__x = 10

a = A()

print(a._A__x)
```

## Composition Concept

```python
class Engine:
    def start(self):
        print("Engine Started")

class Car:
    def __init__(self):
        self.engine = Engine()
```

```python
car = Car()
car.engine.start()
```

## dataclass

```python
from dataclasses import dataclass

@dataclass
class Student:
    name: str
    age: int

s1 = Student("Mohit", 20)
s2 = Student("Mohit", 20)

print(s1);
print(s2);

print(s1 == s2)
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # We need to define the below given functions if we are not using the @dataclass
    def __repr__(self):
        return f"Student(name='{self.name}', age={self.age})"

    def __eq__(self, other):
        return self.name == other.name and self.age == other.age

Students1 = Student(name="Mohit Soni", age=56);
Students2 = Student(name="Mohit Soni", age=56);

print(Students1);
print(Students2);

print(Students1 == Students2);
```

## Slots '''(__slots__)'''

```python
# __slots__ restricts a class to a fixed set of attributes, reducing memory usage and preventing the creation of new attributes.

from dataclasses import dataclass

@dataclass
class Student:
    __slots__ = ("name", "age")

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return str({"name": self.name, "age" : self.age});

s = Student("Mohit", 20)

s.name = "Rahul"      # ✅ Allowed
s.age = 21            # ✅ Allowed

#This __slots__ is used to restrict the entry of the extra unwanted values to associate with the existing class;
# s.city = "Rajkot"     # ❌ AttributeError

print(s)
```

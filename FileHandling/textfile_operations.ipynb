```python
"""
"r" - Read - Default value. Opens a file for reading, error if the file does not exist

"a" - Append - Opens a file for appending, creates the file if it does not exist

"w" - Write - Opens a file for writing, creates the file if it does not exist

"x" - Create - Creates the specified file, returns an error if the file exists
"""
```

```python
import os;
```

```python
path = os.getcwd();
path
```

```python
filename = "testing.txt";

# The write mode of the file overwrite the content of the file;
fileWriter = open(os.path.join(path, filename), "w");

fileWriter.write("This is the newly created file in on the given path");

fileWriter.close();
```

```python
# opening the file in the reader mod and reading the content of the file;

Reader = open(os.path.join(path, filename), "r");
file_content = Reader.read();
Reader.close();
file_content
```

## using with to open a file

```python
# in this way of file handling we do not need to close the file, it automatically closes it 
# after the code block executed;

with open(os.path.join(path, filename), "r") as currfile:
    print(currfile.read())
```

## Append 'a' for the files

```python
with open(os.path.join(path, filename), 'r') as currentfile:
    print("------------------- Read before ------------------");
    print(currentfile.read());

# When we write the file in the append mode instead of overwriting the content, it add the content at the end of the file content;
with open(os.path.join(path, filename), 'a') as currentfile:
   currentfile.write("\nThis is the new line context not overwrite the present content");

with open(os.path.join(path, filename), 'r') as currentfile:
    print("------------------- Read After ------------------");
    print(currentfile.read());
```

### Read Modes
* "r" – Read
* "rt" – Read Text
* "rb" – Read Binary

### Write Modes
* "w" – Write (overwrites)
* "wt" – Write Text
* "wb" – Write Binary

### Append Modes
* "a" – Append
* "at" – Append Text
* "ab" – Append Binary

### Create Mode
* "x" – Create new file

### Read + Write Modes
* "r+"
* "w+"
* "a+"
* "rb+"
* "wb+"
* "ab+"

```python
# The a+ allow us to open file in the append + read mode;

with open(os.path.join(path, filename), "a+") as currentfile:
    currentfile.seek(0);
    print("------------------- Read Before ------------------");
    print(currentfile.read());

    # adding some content to the file;
    currentfile.write("\nNew Content in the file again....");

    # Reading the file content again;
    print("------------------- Read After ------------------");
    currentfile.seek(0) # seek allow us to define the current location of the pointer in the file;
    print(currentfile.read());
```

## seek and tell Function;

```python
with open(os.path.join(path, filename), "r") as currentfile:
    # the tell function allow us to get the current location of the cursor in the file;
    # by default the cursor location in the read mod is 0;
    print(currentfile.tell())
    print("------------------- Reading Content when cursor at 0 ----------------------");
    print(currentfile.read());

    # The seek function allow us to define the location of the cursor in the file;
    currentfile.seek(10);
    print("------------------- Reading Content when cursor at 10 ----------------------");
    print(currentfile.tell()); # pointer at 10;
    print(currentfile.read());
    print(currentfile.tell()); # pointer at the end char of the file;
    # The pointer at the 10 means skip 1st 10 chars;
```

```python
# File methods to get basic details about the file;

with open(os.path.join(path, filename), "a+") as currentfile:
    print(currentfile.mode);
    print(currentfile.encoding);
    print(currentfile.name)
    print(currentfile.closed)
```

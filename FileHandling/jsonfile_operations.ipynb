```python
import json
```

```python
# json model allow us to store the json data and allow to play with the json data


student = {
    "name": "Mohit",
    "age": 20,
    "course": "B.Tech",
    "marks": [85, 90, 95]
}

# json.dump() function, converts the json data to the string and store that data in the targeted file;
with open("student.json", "w") as file:
    json.dump(student, file)
```

```python
import json

with open("student.json", "r") as file:
    # json.load() used to load the json data from the file in the original json form;
    data = json.load(file)

print(data)
print(type(data)) # this data type is dict
print(data["name"])
```

| Function               | Description                              |
| ---------------------- | ---------------------------------------- |
| `json.dump(obj, file)` | Write a Python object to a JSON file     |
| `json.load(file)`      | Read a JSON file into a Python object    |
| `json.dumps(obj)`      | Convert a Python object to a JSON string |
| `json.loads(string)`   | Convert a JSON string to a Python object |

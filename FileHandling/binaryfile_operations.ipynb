| Mode    | Description                               |
| ------- | ----------------------------------------- |
| `"rb"`  | Read binary file                          |
| `"wb"`  | Write binary file (creates or overwrites) |
| `"ab"`  | Append to binary file                     |
| `"rb+"` | Read and write binary file                |
| `"wb+"` | Read and write (overwrites existing file) |
| `"ab+"` | Read and append                           |


```python
# Writing data to the binary file;
data = bytes([65, 66, 67, 68, 69]);

with open("sample.bin", "wb") as file:
    file.write(data);
```

```python
# Reading data from the created binary file;

with open("sample.bin", "rb") as file:
    data = file.read()

print(data)
```

```python
# Reading data byte by byte;
# read takes one optional argument that the no of bytes to read;
# currentfile.read(10) - read the 10 char from the current position of the pointer;

with open("sample.bin", "rb") as file:
    while True:
        byte = file.read(1)
        if not byte:
            break
        print(byte)
```

```python
# Appending data to the binary file we use the ab (append binary);
# if we uses the modes directly ( 'a', 'r', 'w' ) by default they perform operations for the normal text file not the binary file for the binary file, we need to mention explicity;
with open("sample.bin", "ab") as file:
    file.write(b"FGH")
```

```python
# Reading fixed numbers of bytes from the file;

with open("sample.bin", "rb") as file:
    print(file.tell()); # current location of the pointer;
    data = file.read(3) # reading 3 byte from the current location of the pointer;

print(data);
```
